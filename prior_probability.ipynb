```python
f6=open("C:/BDAP/20 - NLP/merged.txt")
arr=[]
text=[]

num_words_female=0

for line in f6:
    arr.append(line)
    words=line.split()
    text.append(words)
    num_words_female += len(words)

#print text

my_dict_f = {}
for item in text:
    for word in item:
        if word in my_dict_f:
            my_dict_f[word] += 1
        else:
            my_dict_f[word] = 1
                 

print num_words_female

#print my_dict

for word in my_dict_f:
    print word,"result",float(my_dict_f[word])/num_words

print len(my_dict_f)
```
